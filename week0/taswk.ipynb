```python
# Problem 3

A = np.array([[3,4],
              [7,8]])

B = np.array([[5,3],
              [2,1]])

A_inv = np.linalg.inv(A)

result = np.dot(A, A_inv)

print("A inverse:\n", A_inv)
print("A * A_inv:\n", np.round(result, 5))
```

```python
AB = np.dot(A,B)
BA = np.dot(B,A)

print("AB:\n", AB)
print("BA:\n", BA)
```

```python
left = np.transpose(AB)

right = np.dot(B.T, A.T)

print("(AB)^T:\n", left)
print("B^T A^T:\n", right)
```
